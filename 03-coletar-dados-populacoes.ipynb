```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/hjort/populacao-municipal/main/"

dados = pd.DataFrame()

for arq in ['estimativa_dou_2014-2021.csv', 'estimativa_dou_2022.csv',
            'estimativa_dou_2023.csv', 'estimativa_dou_2024.csv']:
    
    if dados.empty:
        dados = pd.read_csv(url + arq, sep=';', 
                            dtype={'cod1': 'object', 'cod2': 'object', 'habitantes': 'int'})
    else:
        dados2 = pd.read_csv(url + arq, sep=';',
                             dtype={'cod1': 'object', 'cod2': 'object', 'habitantes': 'int'},
                             names=list(dados.columns.values))
        #dados = dados.append(dados2)
        dados = pd.concat([dados, dados2])

dados['codigo'] = dados[['cod1', 'cod2']].apply(lambda x: int(x[0] + x[1]), axis=1)
dados.drop(['cod1', 'cod2'], axis=1, inplace=True)
dados = dados[['ano', 'codigo', 'habitantes']]
dados
```

```python
df = pd.DataFrame()

for ano in range(dados['ano'].min(), dados['ano'].max() + 1):
    
    dft = dados[dados['ano'] == ano].set_index('codigo')
    dft = dft.drop(['ano'], axis=1).rename({'habitantes': "p%d" % ano}, axis=1)
    
    if df.empty:
        df = dft.copy()
    else:
        df = df.join(dft)

df
```

```python
csv = 'populacoes.csv'
df.to_csv(csv)
```
