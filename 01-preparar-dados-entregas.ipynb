```python
df = pd.read_csv(arq, sep=';', encoding='iso-8859-1')
df
```

```python
dfd = df[['UF', 'Município', 'Ano 2021', 'Ano 2022', 'Ano 2023', 'Ano 2024']].copy()
#dfd
```

```python
dfd = df[['UF', 'Município', 'Ano 2021', 'Ano 2022', 'Ano 2023', 'Ano 2024']].copy()
for c in dfd.columns:
    if c.startswith('Ano'):
        ano = c.split(' ')[-1]
        dfd.rename({c: ano}, inplace=True, axis=1)
        dfd[ano] = dfd[ano].apply(lambda x: int(x.replace('.', '')))
dfd
```

```python
csv = 'entregas-dirpf-2021-2024.csv'
dfd.to_csv(csv, index=False)
```
