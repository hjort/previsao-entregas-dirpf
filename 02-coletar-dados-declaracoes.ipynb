```python
ano = 2021
url = "https://servicos.receita.fazenda.gov.br/publico/EstatisticaIRPF/estatisticasDIRPF_%d.CSV" % ano
arq = "estatisticas-dirpf-%d.csv" % ano

!test -f $arq || wget -O $arq $url
!ls -lah $arq
```

```python
!head $arq
```

```python
df = pd.read_csv(arq, sep=';', encoding='iso-8859-1')
df
```

```python
dfd = df[['UF', 'Município',
          'Ano %d' % (ano - 3),
          'Ano %d' % (ano - 2),
          'Ano %d' % (ano - 1),
          'Ano %d' % ano]].copy()
dfd.rename({'UF': 'uf', 'Município': 'municipio'}, axis=1, inplace=True)
for c in dfd.columns:
    if c.startswith('Ano'):
        nano = "d" + c.split(' ')[-1]
        dfd.rename({c: nano}, inplace=True, axis=1)
        dfd[nano] = dfd[nano].apply(lambda x: int(x.replace('.', '')))
dfd
```

```python
csv = 'declaracoes-%d.csv' % ano
print(csv)
dfd.to_csv(csv, index=False)
```
