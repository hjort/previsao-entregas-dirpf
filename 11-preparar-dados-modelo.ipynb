```python
dfe = pd.DataFrame(excecoes, columns=['codigo', 'uf', 'municipio'])
#dfe
```

```python
dfm = dfm.set_index('codigo').join(dfe.set_index('codigo'), rsuffix='e', how='outer')
dfm['uf'] = dfm['uf'].fillna(dfm['ufe'])
for col in ['municipio', 'municipio2']:
    dfm[col] = dfm[col].fillna(dfm['municipioe'])
dfm['municipio3'] = dfm['municipioe']
for col in ['municipio3']:
    dfm[col] = dfm[col].fillna(dfm['municipio2'])
dfm.drop(['ufe', 'municipioe'], axis=1, inplace=True)
dfm
```

```python
dfp = pd.read_csv('populacoes.csv')
dfp
```

```python
dfd = pd.read_csv('declaracoes.csv')
dfd
```

```python
dfmp = dfm.join(dfp.set_index('codigo'))
dfmp.fillna(0, inplace=True)
for col, dtype in dfmp.dtypes.items():
    if dtype == 'float64':
        dfmp[col] = dfmp[col].astype(np.int32)
dfmp
```

```python
df = dfmp.reset_index().set_index(['uf', 'municipio3']).join(
    dfd.rename({'municipio': 'municipio3'}, axis=1).set_index(['uf', 'municipio3']),
    how='right')
df
```

```python
df = df.reset_index().set_index('codigo')
df
```
